```python
import sys
import pandas as pd
import seaborn 
```

```python
vax_exemptions = pd.read_csv('school_vaccine_exemptions.csv')
elementary_directory = pd.read_csv('elementary_directory.csv', quotechar='"') 
#https://schools.utah.gov/schoolsdirectory

vax_exemptions = vax_exemptions.rename(columns={vax_exemptions.columns[0]: 'School'})
# df = df.dropna(how='all')
vax_exemptions = vax_exemptions.dropna(how = 'all') # this didn't work lol 
```

```python
def classify_school(row):
    if row['Private'] == 'Private':
        return 'Private'
    elif row['Charter'] == 'Charter':
        return 'Charter'
    else: 
        return 'Public'
    

elementary_directory['school_type'] = elementary_directory.apply(classify_school, axis=1)

print(elementary_directory.loc[elementary_directory['school_type'] == 'Private', 'School'])

elementary_directory = elementary_directory.drop(columns = ['Private', 'Charter', 'T-I Prev.', 'T-I Curr.', 'School #', 'District #', 'Link', 'Stats', 'Phone', 'Fax', 'Chartered By', 'Opened', 'Closed', 'Principal/Director', 'Principal Name', 'Principal Email'])
```

```python
def normalize_str(s): # hoping for more schools in common, but this actually didn't really work 
    return s.str.strip().str.lower()

for col in ['School', 'City']:
    elementary_directory[col] = normalize_str(elementary_directory[col])
```

```python
elementary_directory
```

```python
vax_exemptions
```

```python
for col in ['School', 'City']:
    vax_exemptions[col] = normalize_str(vax_exemptions[col])
```

```python
vax_exemptions
```

```python
merged_df = pd.merge(vax_exemptions, elementary_directory, on=['School', 'City'], how='inner')
```

```python
merged_df
```

```python
immunizations = merged_df.drop_duplicates(subset = ['School', 'City'])
immunizations = immunizations.reset_index(drop = True)
```
