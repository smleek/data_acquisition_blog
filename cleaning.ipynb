```python
import sys
import pandas as pd
import seaborn 
```

```python
vax_exemptions = pd.read_csv('school_vaccine_exemptions.csv')
elementary_directory = pd.read_csv('elementary_directory.csv', quotechar='"') 
#https://schools.utah.gov/schoolsdirectory

# Rename the first column (index 0)
vax_exemptions = vax_exemptions.rename(columns={vax_exemptions.columns[0]: 'School'})
# df = df.dropna(how='all')
vax_exemptions = vax_exemptions.dropna(how = 'all') # this didn't work lol 
```

```python
vax_exemptions
```

```python
def normalize_str(s):
    return s.str.strip().str.lower()

for col in ['School', 'City']:
    vax_exemptions[col] = normalize_str(vax_exemptions[col])
    elementary_directory[col] = normalize_str(elementary_directory[col])

merged_df = pd.merge(vax_exemptions, elementary_directory, on=['School', 'City'], how='left')
```

```python
merged_df
```

```python
immunizations = merged_df.drop_duplicates(subset = ['School', 'City'])
immunizations = immunizations.drop(columns = ['Phone', 'Fax', 'Principal Name', 'Conditionally enrolled %', 'Extended conditionally enrolled %', 'Out of compliance %', 'District #', 'School #', 'Grades', 'Link', 'Stats', 'Chartered By', 'Opened', 'Closed', 'Principal/Director', 'Principal Email'])
immunizations = immunizations.reset_index(drop = True)
```

```python
immunizations
```

```python
immunizations['Private'] = immunizations['Private'].isna().astype(int)
immunizations['Charter'] = immunizations['Charter'].notna().astype(int)
immunizations['Public'] = ((immunizations['Private'] == 0) & (immunizations['Charter'] == 0)).astype(int)
```

```python
def classify_school(row):
    if row['Charter'] == 0 and row['Public'] == 1:
        return 'Private'
    elif row['Charter'] == 1 and row['Private'] == 0:
        return 'Public Charter'
    elif row['Charter'] == 1 and row['Private'] == 1:
        return 'Private Charter'
    else:
        return 'Public'
    

immunizations['school_type'] = immunizations.apply(classify_school, axis=1)

print(immunizations.loc[immunizations['school_type'] == 'Public Charter', 'School'])
```
